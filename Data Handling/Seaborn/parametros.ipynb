```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(data=tips, x = 'total_bill');
```

```python
sns.displot(data=tips, x='total_bill', y='tip'); # with kind="type" we can change the histogram to see any other type of histogram
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex');
```

```python
sns.displot(data=tips, x='total_bill', hue='sex');
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde');
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde');
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='kde', palette='dark');
```
