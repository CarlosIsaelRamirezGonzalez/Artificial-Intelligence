```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue="time", stat='frequency');
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue="smoker", stat='probability', multiple='dodge'); # try with stat=percent
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, fill=True);
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat="count");
```
