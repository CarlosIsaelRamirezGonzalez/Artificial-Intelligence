```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000, 800, 900, 1000, 300]
```

```python
plt.bar(country, population, width=.5, color=['red', 'blue']);
plt.ylabel("Population")
plt.xticks(np.arange(5), ('India', 'Japon','Mexico', 'Colombia', 'Bruselas'), rotation=45);
```

```python
plt.barh(country, population);
plt.xlabel("Population")
```
