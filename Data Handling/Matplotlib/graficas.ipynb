```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11) # del 0 al 5 en 11 pasos
y = x ** 2
print(f"x: {x} \n y: {y}")
```

```python
plt.plot(x, y, 'rx')
plt.show()
```

```python
plt.plot(x, y, 'rs-')
plt.show()
```

```python
plt.plot(x, y, 'yD:')
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(y)
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```
