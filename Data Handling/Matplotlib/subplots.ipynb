```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11) # del 0 al 5 en 11 pasos
y = np.sin(x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'b')
axes[1].plot(y, x, 'r');
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y, 'b')
ax2.plot(y, x, 'r');
```

```python
fig, axes = plt.subplots(2, 4) # Filas columnas 
axes[0, 0].plot(x, np.cos(x)) # [filas, columnas]
axes[0, 1].plot(x, np.sin(x), 'y')
axes[0, 2].plot(x, np.tan(x), 'g')
axes[0, 3].plot(x, np.cos(x) ** 2, 'r')
fig.tight_layout();
```
