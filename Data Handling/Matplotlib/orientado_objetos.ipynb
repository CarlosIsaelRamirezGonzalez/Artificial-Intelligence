```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11) # del 0 al 5 en 11 pasos
y = x ** 2
print(f"x: {x} \n y: {y}")
```

```python
fig = plt.figure()
axes = fig.add_axes([.1, .1, .5, .5]) # donde se ubica x, donde se ubica y, ancho, alto
axes.set_facecolor('gray')
axes.plot(x, y, 'b')

axes2 = fig.add_axes([.8, .1, .5, .5])
axes2.plot(y, x, 'r');
```
