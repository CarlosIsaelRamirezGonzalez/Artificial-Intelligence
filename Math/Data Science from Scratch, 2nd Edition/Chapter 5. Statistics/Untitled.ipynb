```python
from typing import List
from collections import Counter

# Tendencias centrales <- Nos da una nocion de donde se concentran los datos
def mean(xs: List[float]) -> float: # El promedio de los datos
    return sum(xs) / len(xs)


def _median_odd(xs: List[float]) -> float:
    return sorted(xs)[len(xs) // 2]

def _median_even(xs: List[float]) -> float:
    sorted_xs = sorted(xs)
    hi_midpoint = len(xs) // 2
    return (sorted_xs[hi_midpoint] + sorted_xs[hi_midpoint - 1]) / 2

def median(v: List[float]) -> float: # El numero de en medio o el promedio de los dos numeros de en medio
    return _median_even(v) if len(v) % 2 == 0 else _median_odd(v)

assert median([1, 10, 2, 9, 5]) == 5
assert median([1, 9, 2, 10]) == (2 + 9) / 2


def quantile(xs: List[float], p: float) -> float:
    p_index = int(.25 * len(xs))
    return sorted(xs)[p_index]

def mode(x: List[float]) -> List[float]: # El nunero o los numeros mas comunes
    counts = Counter(x) #valores clave valor {clave: valor}
    max_count = max(counts.values()) # tupas (clave, valor)
    return [x_i for x_i, count in counts.items() if count == max_count] # x_i es lo que llenara la lista, si y solo si el valor es igual a la mayor cantidad de esa llave 
    
# Resultados de las tendencias centrales:
mean_friends = mean(num_friends)
median_friends = median(num_friends)
q1_friends = quantile(num_friends, .25)
most_common_values = mode(num_friends)
print(f"El promedio de num_friends: {mean_friends}")
print(f"La mediana de num_friends: {median_friends}")
print(f"El cuantil 1 (25%) de num_friends es: {q1_friends}")
print(f"El o los valores más comunes son: {most_common_values}")
```

```python
import math
# Dispersion <- las medidas de que tan dispersos estan los datos

# No se usa range porque ya es una funcion en python
def data_range(xs: List[float]) -> float: # El rango de los datos (El valor que arroja dice mucho, si es cercano a 0 significa que no esta disperso en absoluto, pero si el valor es muy grande significa que hay una gran dispersion entre los valores)               
    return max(xs) - min(xs)


def sum_of_squares(xs: List[float]) -> float:
    xs_mean = mean(xs)
    return sum((x - xs_mean) ** 2 for x in xs)

def de_mean(xs: List[float]) -> List[float]: # se usa para obtener una lista de datos centrados a cero
    x_bar = mean(xs)
    return [x - x_bar for x in xs]

def variance(xs: List[float]) -> float: # Indica cuanto se alejan los valores individuales de la media en un conjunto de datos 
    assert len(xs) >= 2, "la varianza requiere de al menos dos elementos"
    
    n = len(xs)
    deviations = de_mean(xs) # Lista de desviaciones respectoa la media 
    return sum_of_squares(deviations) / (n - 1)

def standar_devitation(xs: List[float]) -> float: # Se usa la desviacion estandar en lugar de la varianza debido a que la varianza tiene unidades al cuadrado, debido a su formula 
    """The standard deviation is the square root of the variance"""
    return math.sqrt(variance(xs))

# El rango intercuartílico (IQR) es una medida más robusta frente a outliers. El rango intercuartílico es la diferencia entre el tercer cuartil (Q3 o 75% -ile) y el primer cuartil (Q1 o 25% -ile) y se calcula utilizando percentiles.
def interquartile_range(xs: List[float]) -> float:
    return quantile(xs, 0.75) - quantile(xs, 0.25)


data_range_num_friends = data_range(num_friends)
standar_devitation_num_friends = standar_devitation(num_friends)
interquartile_range_num_friends = interquartile_range(num_friends)

print(f"El rango de los datos es: {data_range_num_friends}")
print(f"La desviacion estandar es: {standar_devitation_num_friends}")
print(f"El rango intercuartilico es: {interquartile_range_num_friends}")
```

```python

# Correlacion <- Como se relacionan 

# La covarianza no es muy confiable
def covariance(xs: List[float], ys: List[float]) -> float:# La covarianza indica si hay una tendencia general para que ambas variables aumenten o disminuyan al mismo tiempo  
    assert len(xs) == len(ys), "xs and ys must have same number of elements"
    xs_mean = mean(xs)
    ys_mean = mean(ys)
    sum_ys_xs = sum((x - xs_mean) * (y - ys_mean) for x, y in zip(xs, ys)) 
    return sum_ys_xs / len(xs)

def correlation(xs: List[float], ys: List[float]) -> float:
    stdev_x = standar_devitation(xs)
    stdev_y = standar_devitation(ys)
    if stdev_x > 0 and stdev_y > 0:
        return covariance(xs, ys) / stdev_x / stdev_y
    else: # No se puede calcular con esta formula, por ende es 0 o indefinida
        return 0 #  "si una de las variables no tiene variación (todos los valores son iguales), la correlación es indefinida o no tiene un valor significativo".

cov_dataSciencester_VP = covariance(num_friends, daily_minutes)
correlation_dataSciencester_VP_minutes = correlation(num_friends, daily_minutes)
correlation_dataSciencester_VP_hours = correlation(num_friends, daily_hours)
print(f"La covarianza: {cov_dataSciencester_VP}")
print(f"Correlacion amigo-minuto: {correlation_dataSciencester_VP_minutes}")
print(f"Correlacion amigo-hora: {correlation_dataSciencester_VP_hours}")
```
