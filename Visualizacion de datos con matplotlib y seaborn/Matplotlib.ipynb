# Subplot

Subplot sirve para graficar diversos datos de manera ordenada sin que se afecten entre si, si intentamos graficar varios graficos en una sola celda de codigo, lo normal es que todos estos graficos se nos amontonen, con subplot esto es otra historia. 

```python
filas=1
columnas=2
index=1
plt.subplot(filas,columnas,index)
plt.plot(x,y,'rx')
index+=1
plt.subplot(filas,columnas,index)
plt.hist(x)
plt.show()
```

# Metodo orientado a objeto

```python
figura=plt.figure()#Creamos una figura
axes=figura.add_axes([10,10,0.5,0.7])#Los dos primeros parametros hacen conexion de la figura con el lienzo
#Los siguientes dos parametros se encargan del ancho y la altura respectivamente
axes.plot(x,y,'b')
plt.show()
```

Es posible que se piense que lo que acabamos de hacer es lo mismo que se puede hacer con pyplot, pero hay mas que eso, veamos a continuacion mas

```python
figura=plt.figure()#Creamos una figura

axes=figura.add_axes([0.1,0.1,0.8,0.9])
axes2=figura.add_axes([0.2,0.55,.4,.3])
#el 0.1,0.1 en axes representa la ubicacion del grafico, por eso se encima uno sobre el otro, pues
#hemos acomodado el axes2 de tal manera que se ponga en un punto que no estorbe al axes

axes2.plot(y,x,'r')
axes.plot(x,y,'b')

axes2.set_facecolor('gray')

plt.show()
```

Explicacion del codigo-> Ahora hemos visto mejor el comportamiento de la grafica orientada a objetos, imaginemos que figura=plt.figure es el constructor, y axes,axes2 los objetos, y .plot,.set_facecolor los metodos, y axes.plot solo son los objetos interactuando con los metodos, los cambios que hagamos no afectan al otro en lo absoluto

# Subplots

Lo mismo que subplot, pero ahora juntaremos el metodo orientado a objetos y el poder de subplot

```python
X=np.linspace(1,10,10)
fig,axes=plt.subplots(nrows=2,ncols=4)
fig.tight_layout()#Acomoda mejor el tamao de cada grafica 
axes[0,0].plot(x,np.sin(x),'#264653')
axes[0,1].plot(x,np.cos(x),'#606c38')
axes[0,2].plot(x,np.tan(x),'#cd9777')
axes[0,3].plot(x,-np.arccos(x),'#b7b7a4')
axes[1,0].plot(x,np.arccos(x),'#fca311')
axes[1,1].plot(x,np.arcsin(x),'#cce3de')
axes[1,2].plot(x,-np.sin(x),'#0d1b2a')
axes[1,3].plot(x,np.tan(x),'#ddb892')
```

Podemos tener el mismo resultado de la siguiente manera

```python
fig,((ax0,ax1,ax2,ax3),(ax4,ax5,ax6,ax7))= plt.subplots(nrows=2,ncols=4)
ax1.plot(x,np.sin(x),'#264653')
ax2.plot(x,np.cos(x),'#606c38')
ax3.plot(x,np.tan(x),'#cd9777')
ax4.plot(x,-np.arccos(x),'#b7b7a4')
ax5.plot(x,np.arccos(x),'#fca311')
ax6.plot(x,np.arcsin(x),'#cce3de')
ax7.plot(x,-np.sin(x),'#0d1b2a')
```

# Leyendas, etiquetas, titulos, tamaos

```python
x=np.linspace(0,5,11)
y=np.sin(x)
```

```python
fig,axes=plt.subplots(1,2)
fig.tight_layout()
axes[0].plot(x,y,label='sin(x)')
axes[1].plot(y,x,label='sin(x)')

axes[0].set_title('Relacion X-Y')#Pone titulo
axes[1].set_title('Relacion Y-X')

axes[0].set_xlabel('X')#Pone una etiqueta en el eje x
axes[0].set_ylabel('Y')#Pone una etiqueta en el eje y
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')

axes[0].legend()#Mostramos la leyenda, la leyenda fue declarada en axes[0].plot(x,y,label='sin(x)')
axes[1].legend(loc='lower left')#le indicamos donde queremos la leyenda
axes[1].legend(bbox_to_anchor=(0.5,0.2))#Una manera mas personalizada de poner la ubicacion de la leyenda
```

# Colores y estilos

```python
x=np.linspace(0,5,11)
plt.style.available#Nos manda una lista con todos los estilos disponibles
plt.style.use('dark_background')


fig,ax=plt.subplots()
ax.plot(x,x+1,'r',alpha=0.3,linewidth=2,linestyle=':',marker='o',markersize=5,markerfacecolor='#264653')
ax.plot(x,x+2,'g',alpha=0.6,linewidth=4,linestyle='-',marker='v',markersize=10,markerfacecolor='#2a9d8f')
ax.plot(x,x+3,'b',alpha=0.9,linewidth=6,linestyle='--',marker='X',markersize=15,markerfacecolor='#e9c46a')
```

alpha = Que tan borrosa es la linea.
linewidth = El grosor de la linea.
linestyle = El estilo de la linea.
marker = Marca los puntos donde colicionan x y y.
markersize = El tamao del marker.
markerfacecolor = El color de el marcador.

# Bar Plot

Nos permite graficar variables categoricas

```python
Paises=['Japan','Mexico','Germany','Brazil']
Poblacion=[3000,2000,10000,12000]
plt.style.use('classic')
plt.bar(Paises,Poblacion,width=0.5,color=['red','blue'])#Width es la distancia entre las graficas
plt.show()
```

Tambien podemos modificar como se ven las variables categoricas

```python
Paises=['Japan','Mexico','Germany','Brazil']
Poblacion=[3000,2000,10000,12000]
plt.style.use('classic')
plt.bar(Paises,Poblacion,width=0.5,color=['red','blue'])
plt.xticks(np.arange(4),('Japon','Mexico','Alemania','Brasil'),rotation=45)
plt.show()
```
