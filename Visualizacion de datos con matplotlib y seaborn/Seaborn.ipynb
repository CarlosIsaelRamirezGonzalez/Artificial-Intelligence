# Parametros mas usados en seaborn

seaborn tiene una variedad enorme de graficos, graficos que pueden cambiar completamente si le agregamos un parametro mas.

```python
dtf_tips=sns.load_dataset('tips')
```

```python
sns.displot(data=dtf_tips,x='total_bill')
plt.show()
```

Al poner dos variables dentro de displot el histograma dejara de funcionar para visualizar los datos de manera correcta, por eso mismo lo mejor seria usar un diagrama distinto, seaborn lo elige automaticamente al algregarle una segunda variable

```python
sns.displot(data=dtf_tips,x='total_bill',y='tip')
plt.show()
```

al agregarle hue lo que hara sera hacer una separacion, en este caso la haremos por sexos

```python
sns.displot(data=dtf_tips,x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
sns.displot(data=dtf_tips,x='total_bill',hue='sex')
plt.show()
```

Tambien podemos cambiar el tipo de grafica que estamos viendo a traves de un parametro, y ese es kind

```python
sns.displot(data=dtf_tips,x='total_bill',hue='sex',kind='kde',legend=True,palette='dark',alpha=.9)
plt.show()
```

# Distribuciones

Seaborn ofrece algunas gr√°ficas u opciones que te permiten trabajar con distribuciones orientadas a tipos de datos num√©ricos.

```python
sns.histplot(data=dtf_tips,x='tip',bins=15)#Bins es el numero de barras que queremos mostrar
plt.show()
```

```python
sns.histplot(data=dtf_tips,x='tip',bins=15,cumulative=True)#Acumula los valores
plt.show()
```

```python
sns.histplot(data=dtf_tips,x='tip',bins=15,cumulative=False,hue='sex',stat='frequency')#Trabaja con la frecuencia
plt.show()
```

```python
sns.histplot(data=dtf_tips,x='tip',bins=15,cumulative=False,hue='sex',stat='probability')
plt.show()
```

stat='density'
stat='percentil'
stat='count' <- default

```python
sns.histplot(data=dtf_tips,x='tip',bins=15,cumulative=False,hue='sex',stat='probability',multiple='dodge')
plt.show()
```

multiple='stack' <- Uno encima del otro   multiple='layer' <- Default    multiple='dodge' <- Uno encima del otro    multiple='fill'

Diagrama de densidad

```python
sns.kdeplot(dtf_tips,x='tip',hue='sex',cumulative=True)
plt.show()
sns.kdeplot(dtf_tips,x='tip',hue='sex',fill=True)#Diagrama de densidad
plt.show()
```

Grafico escalonado

```python
sns.ecdfplot(data=dtf_tips,x='tip',hue='sex')
plt.show()
sns.ecdfplot(data=dtf_tips,x='tip',hue='sex',stat='count')#Con stat='count' vemos cuantos 
plt.show()
```

Todo lo que acabamos de programar lo podemos hacer con una sola funcion y agregandole un parametro.

```python
sns.displot(data=dtf_tips,x='tip',hue='sex',kind='hist')
plt.show()
```

kind='hist' <- Histograma
kind='kde'  <- Grafico de densidad
kind='ecdf' <- Grafico escalonado

# Categoricos

Seaborn ofrece diferentes gr√°ficos para graficar datos categ√≥ricos o variables de texto. 

```python
sns.countplot(data=dtf_tips,x='day',hue='sex')#Cuenta la aparicion
plt.show()
```

```python
sns.stripplot(data=dtf_tips,x='day',y='total_bill',hue='sex')#Analiza la distribucion respecto a otra variable
plt.show()
#Si queremos separar los putos debemos poner como parametro dodge=True
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=dtf_tips,x='day',y='total_bill',hue='sex')#No encima los puntos(Datos)
plt.show()
#Si queremos separar los putos debemos poner como parametro dodge=True
```

```python
sns.violinplot(data=dtf_tips,x='day',y='total_bill',hue='sex',split=True,palette='pastel')
plt.show()
#Nos muestra la distribucion de nuestros datos de con forma de violin
```

```python
sns.catplot(data=dtf_tips,x='day',y='total_bill',hue='sex')
plt.show()
```

```python
sns.catplot(data=dtf_tips,x='day',y='total_bill',hue='sex',kind='swarm');
```

```python
sns.catplot(data=dtf_tips,x='day',y='total_bill',hue='sex',col='time');
```

```python
sns.catplot(data=dtf_tips,x='day',y='total_bill',hue='sex',kind='swarm');
```

# Relation

Seaborn maneja los gr√°ficos de relaci√≥n entre distintas variables num√©ricas para hacer visualizaciones.

Relacion a traves de un grafico de puntos

```python
plt.figure(figsize=(8,8))
markers={"Lunch":"D","Dinner":"s"}
sns.scatterplot(data=dtf_tips,x='total_bill',y='tip',hue='day',style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,.5));
#Acomodamos la leyenda
```

```python
sns.relplot(data=dtf_tips,x='total_bill',y='tip',hue='day',style='time',size='size',col='time');
#Lo mismo con los cambios, solo con kind.
```

# Joinplot/Pairplot

Joinplot

```python
sns.jointplot(data=dtf_tips,x='total_bill',y='tip');
```

```python
sns.jointplot(data=dtf_tips,x='total_bill',y='tip',hue='sex');#Separamos por sexo
```

```python
sns.jointplot(data=dtf_tips,x='total_bill',y='tip',hue='sex',kind='hist');#Cambiamos a histograma
```

Tambien podemos modificar la parte del histograma sin afectar el centro

```python
sns.jointplot(data=dtf_tips,x='total_bill',y='tip',hue='sex',kind='hist'
              ,marginal_ticks=True,
              marginal_kws=dict(bins=25,fill=False,multiple='dodge'));#Cambiamos a histograma
```

Pairplot

```python
sns.pairplot(data=dtf_tips);#Tomara total_bill, tip y size
#La especialidad del pairplot es la relacion entre variables numericas
```

Tambien podemos hacer que separe por sexo

```python
sns.pairplot(data=dtf_tips,hue='sex',palette='coolwarm');
```

```python
sns.pairplot(data=dtf_tips,hue='sex',palette='coolwarm',corner=True);#No muestra datos repetidos
```
