```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Preparacion de los datos 

```python
x=np.linspace(1,100,10)
y=x**2
```

# Graficacion de los datos

Hay varias formas de poder graficar un conjunto de datos, si solo necesitas una grafica te recomiendo

```python
plt.plot(x,y)
plt.show
```

Sin embargo, si necesitas mas graficas lo mejor es subplots, hay 3 maneras de graficar muchas graficas, subplot, orientado a objetos, y subplots, imagina subplot para graficar de 1-2, orientado a objetos 2-Adelante, pero conforme las graficas aumenten, tu codigo lo hara,  y sin duda te perderas, una alternativa mejor es subplots, que combina lo mejor de las anterior mencionadas, ademas que es rapido, facil de entender, y tu codigo se vera mas limpio.

A continuacion vamos a graficar 8 graficas en menos de 12 lineas de codigo: 

```python
x=np.linspace(1,10,10)
fig,axes=plt.subplots(2,4)
fig.tight_layout()#Acomoda mejor el tamao de cada grafica 
axes[0,0].plot(x,(x**3),'#264653')
axes[0,1].plot(x,np.cos(x),'#606c38')
axes[0,2].plot(x,(x**2),'#cd9777')
axes[0,3].plot(x,-np.tan(x),'#b7b7a4')
axes[1,0].plot(x,(x//2),'#fca311')
axes[1,1].plot(x,np.sin(x),'#cce3de')
axes[1,2].plot(x,-np.sin(x),'#0d1b2a')
axes[1,3].plot(x,np.tan(x),'#ddb892')
```

# Agregando detalles

Un punto importante a la hora de saber como graficar son los detalles,  de esta manera es posible que los resultados de los graficos sean mas claros, y no caigan en algun tipo de ambiguedad.

```python
#Ignorar codigo a partir de aqui
fig,axes=plt.subplots(2,4)
fig.tight_layout()
axes[0,0].plot(x,(x**3),'#264653',label='x^3')
axes[0,1].plot(x,np.cos(x),'#606c38',label='cos(x)')
axes[0,2].plot(x,(x**2),'#cd9777',label='x^2')
axes[0,3].plot(x,-np.tan(x),'#b7b7a4',label='-tan(x)')
axes[1,0].plot(x,(x//2),'#fca311',label='x//2')
axes[1,1].plot(x,np.sin(x),'#cce3de',label='sen(x)')
axes[1,2].plot(x,-np.sin(x),'#0d1b2a',label='-sen(x)')
axes[1,3].plot(x,np.tan(x),'#ddb892',label='-tan(x)')
#Hasta aqui.

#Poner titulos
axes[0,0].set_title('x~x**3')
axes[0,1].set_title('x~cos(x)')
axes[0,2].set_title('x~x**2')
axes[0,3].set_title('x~-tan(x)')
axes[1,0].set_title('x~x//2')
axes[1,1].set_title('x~sen(x)')
axes[1,2].set_title('x~-sen(x)')
axes[1,3].set_title('x~tan(x)')

#Mostrar leyendas
axes[0,0].legend()
axes[0,1].legend()
axes[0,2].legend()
axes[0,3].legend()
axes[1,0].legend()
axes[1,1].legend()
axes[1,2].legend()
axes[1,3].legend()

#Etiquetas
axes[0,0].set_xlabel('X')
axes[0,0].set_ylabel('Y')
axes[0,1].set_xlabel('X')
axes[0,1].set_ylabel('Y')
axes[0,2].set_xlabel('X')
axes[0,2].set_ylabel('Y')
axes[0,3].set_xlabel('X')
axes[0,3].set_ylabel('Y')
axes[1,0].set_xlabel('X')
axes[1,0].set_ylabel('Y')
axes[1,1].set_xlabel('X')
axes[1,1].set_ylabel('Y')
axes[1,2].set_xlabel('X')
axes[1,2].set_ylabel('Y')
axes[1,3].set_xlabel('X')
axes[1,3].set_ylabel('Y')
```

En cuestion de leyendas se pueden ubicar donde nosotros queramos, en este caso no fue asi, debido al poco tamao de cada grafica, pero el acomodo es el siguiente:
.legend(loc='lower left') -> Acomoda abajo a la izquierda
o si quieres ponerlo a traves de coordenadas de la siguiente manera:
.legend(bbox_to_anchor=(0.5,0.2))
