```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dtf_tips=sns.load_dataset('tips')
dtf_tips.head(2)
```

# Distribuciones

Histograma

```python
sns.displot(data=dtf_tips,x='size',hue='sex',kind='hist',bins=10,
            cumulative=False,stat='frequency',multiple='dodge');
sns.displot(data=dtf_tips,x='size',hue='sex',kind='hist',bins=10,
            cumulative=True,stat='probability',multiple='dodge');
```

Funciones usadas:

Diagrama de densidad

```python
sns.displot(data=dtf_tips,x='size',kind='kde',hue='sex',cumulative=True,fill=True);
sns.displot(data=dtf_tips,x='size',kind='kde',hue='sex',fill=True);
```

Funciones usadas:

Grafico escalonado

```python
sns.displot(data=dtf_tips,x='size',hue='sex',stat='count',kind='ecdf');
sns.displot(data=dtf_tips,x='size',hue='sex',kind='ecdf');
```

Funciones usadas:

data=(El data frame)
hue=(La variable que queremos que se use como separacion de los datos)
kind=(El tipo de grafica)
stat=(La distribucion estadistica que querramos[proportion,count])    

# Variables Categoricas

Seaborn nos ofrece una herramienta muy poderosa a la hora de querer graficar variables categoricas, esta es catplot. catplot lo que nos permite es poner en su eje x o y alguna variable categorica, y en su eje contrario una variable numerica, de esta forma podemos visuzliar la concurrencia de los datos

```python
sns.catplot(data=dtf_tips,x='day',y='total_bill',col='time',hue='smoker',kind='strip')
```

# Relaciones

Cuando estemos haciendo algun tipo de investigacion estadistica deberemos vernos en la tarea de armar ciertas relaciones entre variables, para descubrir estas relaciones es necesario verlas en distintos graficos, seaborn nos ofrece diversas herramientas para esta tarea

Grafico de puntos

```python
sns.relplot(data=dtf_tips,x='total_bill',y='tip',hue='day',style='time',size='size',col='smoker');
```

Funciones usadas:

Joinplot

```python
sns.jointplot(data=dtf_tips,x='total_bill',y='tip',hue='sex',kind='hist'
              ,marginal_ticks=True,
              marginal_kws=dict(bins=25,fill=False,multiple='dodge'));#Cambiamos a histograma
```

Funciones usadas:

data=(El data frame)
hue=(La separacion)
kind=(El tipo de grafica que queremos[kde,hist,hex,reg,resid,scatter])
marginal_ticks=(Si es falso, suprimira las marcas en el eje de conteo/densidad de las tablas externas)
marginal_kws=(Acepta diccionarios como parametro, es la forma en la que podemos interactuar con el grafico externo)

Pairplot

```python
sns.pairplot(data=dtf_tips,hue='sex',palette='coolwarm',corner=True,kind='scatter',diag_kind='auto')
```

Funciones usadas

data=(El data frame)
hue=(La separacion)
palette=(la paleta de colores que queremos usar[coolwarm,rocket,mako,flare,crest,magma,viridis])
corner=(De usar falso, mostrara todos los valores repetidos)
kind=(El tipo de grafica que queremos[scatter,kde,hist,reg])
diag_kind=(La grafica que queremos en el eje diagonal[auto,hist,kde,None])

Heatmap

```python
sns.heatmap(dtf_tips.corr(numeric_only=True),annot=True,cmap='coolwarm',
            linewidths=0.07,linecolor='black',vmin=0,vmax=1)
```

Funciones usadas:

dtf_tips.corr(numeric_only=True)={Nos manda las correlaciones de todas nuestras variables numericas}
annot=(De ser verdadero nos mostrara el valor de todas las correlaciones en los respectivos cuadrados)
cmap=(la paleta de colores que queremos usar[coolwarm,rocket,mako,flare,crest,magma,viridis])
linewidths=(El tamao de las lineas)
linecolor=(El color de las lineas)
vmin=(El valor que pongamos lo tomara como minimo y lo mostrara de esa manera en la tabla derecha )
vmax=(El valor que pongamos lo tomara como maximo y lo mostrara de esa manera en la tabla derecha)
