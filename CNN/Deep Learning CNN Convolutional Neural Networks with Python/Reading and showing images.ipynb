```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = plt.imread('D:\IA\CNN\Deep Learning CNN Convolutional Neural Networks with Python\imgs\image.jpg')
```

```python
R = im[:, :, 0]
G = im[:, :, 1]
B = im[:, :, 2]
```

```python
plt.imshow(R, cmap='gray')
plt.axis('off');
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 15))
# axis 1
ax1.set_title('Red')
ax1.imshow(R, cmap='gray')
ax1.axis('off')
# axis 2
ax2.set_title('Green')
ax2.imshow(G, cmap='gray')
ax2.axis('off')
# axis 3
ax3.set_title('Blue')
ax3.imshow(B, cmap='gray')
ax3.axis('off')
# axis 4
ax4.set_title('3 channels')
ax4.imshow(im, cmap='gray')
ax4.axis('off');
```

```python
im = plt.imread('D:\IA\CNN\Deep Learning CNN Convolutional Neural Networks with Python\imgs\image.jpg')
```

```python
im2 = im[ :, :, :3 ]
```

```python
R = im2[ :, :, 0 ]
G = im2[ :, :, 1 ]
B = im2[ :, :, 2 ]
```
